🎯 **Metodo recomendado: Chi-cuadrado (Chi-Square Test)**

Nos sirve para comprobar la independencia o asociación entre variables categóricas, ayudando a identificar patrones y relaciones que pueden ser relevantes para la toma de decisiones o el modelado predictivo.

- El Chi-Cuadrado Test mide la asociación entre variables categóricas y la variable objetivo (Churn).
- p-value muy bajo (< 0.05) → indica que la variable está significativamente asociada con Churn.
- p-value alto (> 0.05) → sugiere que la variable no aporta información relevante y podría eliminarse



```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import chi2

# Crea copia
df_rank = df.copy()

# Convertir Churn para binario
df_rank["Churn"] = df_rank["Churn"].map({"Sí": 1, "No": 0})

# Separar variables categóricas
variables_categoricas = df_rank.select_dtypes(include=["object"]).columns

# Aplicar Label Encoding temporario
le = LabelEncoder()
for col in variables_categoricas:
    df_rank[col] = le.fit_transform(df_rank[col])

# Separar X e y
X_cat = df_rank[variables_categoricas]
y = df_rank["Churn"]

# Aplicar teste Qui-cuadrado
chi_scores, p_values = chi2(X_cat, y)

# Criar DataFrame com ranking
chi_df = pd.DataFrame({
    "Variable": variables_categoricas,
    "Chi2 Score": chi_scores,
    "p-value": p_values
})

# Ordenar por importancia
chi_df = chi_df.sort_values(by="Chi2 Score", ascending=False)

# Mostrar ranking completo
print("//****************************************************//")
print("Ranking de variables categóricas respecto a Churn:")
print("//****************************************************//")

print(chi_df)
print("//****************************************************//")

# Filtrar las menos significativas (p-value > 0.05)
variables_a_eliminar = chi_df[chi_df["p-value"] > 0.05]["Variable"].tolist()
print("\n\n//------------------------------------------------------------------//")
print("\033[1mVariables menos significativas (candidatas a eliminar):\033[1m")
print("\033[1m***", variables_a_eliminar, "***\033[0m")
print("//------------------------------------------------------------------//")
```

```python
variables_significativas = chi_df[chi_df["p-value"] < 0.05]

print("\n✅ Variables significativas (p < 0.05):")
display(variables_significativas)
```

```python
variables_significativas = chi_df[chi_df["p-value"] > 0.05]

print("\n✅ Variables a Eliminar por evita multicolinealidad:")
display(variables_significativas)
```

## 🎯 Split -1 - Separando el Dataset ("X" e "y")

- X (variables predictoras): contiene todas las demás características que explican o influyen en la cancelación.
- y (variable objetivo): contiene la información que queremos predecir (en este caso, si el cliente cancela o no).

_____________________________________________________________________________



```python
X = df.drop("Churn", axis=1)
y = df["Churn"].map({"Sí": 1, "No": 0})
```

**Separar "X" y "y", es fundamental porque?**

** Evita transformar la variable objetivo: **
  - El preprocesamiento debe aplicarse solo a las variables explicativas.
  - Incluir Churn en el proceso de codificación, se estaría alterando la variable que se quiere predecir, lo cual rompe el modelo.

** Claridad en el flujo de trabajo: **
  - X se usa para entrenar el modelo.
  - y se usa como etiqueta para supervisar el aprendizaje.
  - Esto permite aplicar transformaciones solo a X (ej. escalado, encoding, imputación de valores).

** Compatibilidad con librerías de machine learning: **
  - Funciones como train_test_split, fit, predict en scikit-learn esperan que "X" y "y" estén separados.
  - "X" debe ser una matriz de características, y "y" un vector con la variable objetivo.

_______________________________________________________________________________


**En resumen:** Separar X y y antes del One-Hot Encoding asegura que:
- La variable objetivo no se distorsione.
- El preprocesamiento se aplique solo a las variables explicativas.
- El modelo pueda entrenarse correctamente con X y evaluar con y.

**4.  Modelo 4 — Regresión Logística con balanceo y SMOTE**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

# 1. Separar X e y
X = df.drop("Churn", axis=1)
y = df["Churn"].map({"Sí": 1, "No": 0})

# 2. Aplicar One-Hot Encoding a X (antes de dividir en train/test)
X = pd.get_dummies(X, drop_first=True)

# 3. Dividir en entrenamiento y prueba (estratificado)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 4. Balancear clases con SMOTE (oversampling de la clase minoritaria)
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# 5. Escalar variables numéricas
scaler = StandardScaler()
X_train_res_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test)

# 6. Entrenar modelo de regresión logística con class_weight balanced
log_model = LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42)
log_model.fit(X_train_res_scaled, y_train_res)

# 7. Predicciones
y_pred = log_model.predict(X_test_scaled)

# 8. Evaluación
print("Regresión Logística con balanceo y SMOTE")
print(classification_report(y_test, y_pred))

# 9. Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
print("Matriz de Confusión:\n", cm)
```

##🎯Factores que más influyen en la cancelación

-	El estudio confirma que el churn (fuga de clientes) ocurre principalmente en etapas tempranas de la relación contractual, con una mediana de apenas 10 meses frente a 29 meses en clientes que permanecen.

-	Variables como tipo de internet (fibra óptica), uso de servicios de streaming y contratos mensuales incrementan el riesgo de abandono, mientras que la antigüedad del contrato y mayor gasto acumulado actúan como factores protectores.

-	Los modelos de regresión logística muestran que, sin balanceo, el desempeño es bueno para la clase mayoritaria (“No”), pero insuficiente para detectar la clase minoritaria (“Sí”).

-	Al aplicar SMOTE y class_weight balanced, el modelo mejora el recall de la clase “Sí” (0.69), logrando identificar más clientes en riesgo, aunque con menor precisión (0.54).

-	La matriz de confusión y los reportes de métricas confirman que el modelo balanceado ofrece un mejor equilibrio entre ambas clases, reduciendo falsos negativos en churn.

## 🎯 Recomendaciones - Estrategias de retención basadas en los resultados obtenidos.

La estrategia debe centrarse en detectar tempranamente clientes vulnerables, reforzar la experiencia inicial, incentivar contratos largos y pagos automáticos, y aplicar el modelo predictivo balanceado con SMOTE para maximizar la retención.

**Propuestas de accionables:**

1. Refuerzo en etapas tempranas y contratos largos
    - Los clientes con contratos mensuales y antigüedad < 10 meses son los más propensos al churn.
    - Estrategias: programas de onboarding proactivo, incentivos para contratos anuales/bianuales, y monitoreo de la calidad del servicio de fibra óptica.
2. Optimización de métodos de pago
    - Los pagos manuales (cheque electrónico o por correo) se asocian con mayor churn.
    - Estrategias: promover pagos automáticos con incentivos y enviar alertas proactivas a clientes con pagos manuales.
3. Protección de segmentos vulnerables
    - Adultos mayores (+60), clientes sin pareja y quienes no tienen soporte técnico son más propensos a cancelar
    - Estrategias: soporte técnico personalizado, canales de atención simplificados y paquetes de valor adicional (seguridad digital, asistencia premium).
4. Gestión del valor percibido y servicios de streaming
    - Clientes con mayor gasto mensual son más leales, mientras que quienes usan streaming presentan mayor riesgo de churn.
    - Estrategias: programas de fidelización para clientes de alto valor y análisis del uso de streaming para ajustar la oferta o mejorar la experiencia.
5. Modelo predictivo recomendado
    - Regresión Logística con balanceo y SMOTE: mejora el recall de la clase “Sí” (0.69), permitiendo identificar más clientes en riesgo.
    - Ventajas: interpretabilidad de coeficientes y efectividad en detección de churners, incluso con algunos falsos positivos aceptables frente al costo de perder clientes.

##🎯Conclusión
Basándonos en los resultados obtenidos, el **Modelo de Regresión Logística con balanceo de clases mediante SMOTE** es el más recomendable para predecir la cancelación de clientes que ocurre en clientes de corta permanencia y con ciertos servicios asociados, ya que mejora la detección de clientes en riesgo, nos entrega un enfoque que maximiza la capacidad de retención y permite diseñar estrategias de retención más efectivas.


**Justificación:**

En un contexto de predicción de churn, es crucial maximizar la detección de los clientes que sí van a cancelar (recall de la clase 'Sí') para poder intervenir con estrategias de retención. Aunque el modelo de Regresión Logística balanceado mostró una precisión ligeramente menor para la clase 'Sí' en comparación con el modelo sin balanceo (0.54 vs 0.65), la mejora significativa en el recall (0.69 vs 0.52) es fundamental. Esto significa que el **modelo balanceado SMOTE** es capaz de identificar a más clientes que realmente abandonarán, lo cual es el objetivo principal de este tipo de predicción.



Además, la Regresión Logística ofrece la ventaja de que sus coeficientes son directamente interpretables, lo que permite entender qué factores específicos (como tipo_internet_Fibra optica, valor_mensual, tiempo_contrato) influyen más en la probabilidad de churn, facilitando el diseño de estrategias de retención focalizadas.
